```python
import pandas as pd
import uuid
import json

import os
```

```python
EXCEL_FILE = '/Users/pabloandres/Documents/GitHub/Survey_MVP_IETS/data/MapeoEntidades_BLH.xlsx'
```

```python
df = pd.read_excel(EXCEL_FILE)

COLUMNA_IPS = 'Organización'  # Cambia si tu columna tiene otro nombre

# Filtra filas válidas (por si hay vacíos)
df = df[df[COLUMNA_IPS].notnull()]
df
```

```python
# Genera hashes únicos por cada IPS
ips_hash_map = {}
for org in df[COLUMNA_IPS]:
    # Quita espacios al inicio/final y asegúrate de string
    org_clean = str(org).strip()
    hash_id = str(uuid.uuid4())
    ips_hash_map[hash_id] = org_clean
```

```python
ips_hash_map
```

```python
# Guarda como JSON
with open('/Users/pabloandres/Documents/GitHub/Survey_MVP_IETS/data/ips_lookup.json', 'w', encoding='utf-8') as f:
    json.dump(ips_hash_map, f, indent=2, ensure_ascii=False)

print(f"Archivo 'ips_lookup.json' creado con {len(ips_hash_map)} IPS.")
```

```python
# Guarda ips_hash_map como Excel en el mismo directorio del JSON
excel_path = os.path.splitext(f.name)[0] + '.xlsx'
pd.DataFrame(list(ips_hash_map.items()), columns=['Hash_ID', COLUMNA_IPS]).to_excel(excel_path, index=False)
print(f"Archivo Excel guardado en: {excel_path}")
```
